```python
import numpy as np
import pandas as pd
from bokeh.layouts import gridplot
from bokeh.plotting import figure, output_file, show
from bokeh.palettes import Spectral4
import time

time1 = time.time()
print("Importing")
LEU = pd.read_excel("Total_Core.xlsx", "LEU")
NU = pd.read_excel("Total_Core.xlsx", "NU")
Total = pd.read_excel("Total_Core.xlsx", "Total")
print("Imported, Importing took {:.1f} Seconds".format(time.time()-time1))
```

```python
TOOLS = "pan,xwheel_zoom, ywheel_zoom,box_zoom,reset,save,box_select"
p = figure(plot_width=800, plot_height=800, sizing_mode='stretch_both',\
           tools = TOOLS, \
           y_range=DataRange1d(bounds = (0,20)), \
           x_range = DataRange1d(bounds = (-1, 1.2* Total["Unnamed: 0"].iloc[-1])))#, x_axis_type="datetime")
p.title.text = 'Click on legend entries to hide the corresponding lines'
print("Processing Data")
elements = ["mo99", "xe133", "i133","tc99"]
time2 = time.time()
for data, name in zip([LEU, NU, Total], ["LEU", "NU", "Total"]):
# for data, name, color in zip([Total], ["Total"], Spectral4):
    df = pd.DataFrame(data)
#     df.index = pd.to_datetime(df.index)
    for ele, color in zip(elements,Spectral4):
        p.line(df["Unnamed: 0"], df[ele], line_width=2, color=color, alpha=0.8, legend_label=name + " " + ele)


p.legend.location = "top_left"
p.legend.click_policy="hide"
p.xaxis.axis_label = 'Days'
p.yaxis.axis_label = 'Activity of Mo-99 (Ci)'
print("Preparing HTML file, Processing took {:.1f} seconds".format(time.time()-time2))
time3 = time.time()
output_file("Total_Core_Interactive_Plot.html", title="Total Core Interactive Plot")
print("HTML Prep took {:.1f} seconds".format(time.time()-time3))
print("Total time to run was {:.1f} seconds".format(time.time()-time1))
show(p)
```
