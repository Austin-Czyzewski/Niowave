```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
import time
from IPython.display import display, clear_output
import glob
import os
from scipy.interpolate import InterpolatedUnivariateSpline, UnivariateSpline, splrep, splev
import scipy.signal
```

```python
Files = glob.glob("*EC.txt")
Open_Files = []
for file in range(len(Files)):
    Temp_file = open(Files[file],'r')
    for i in Temp_file:
        Open_Files.append(i.strip("())\n"))
    Temp_file.close()
for line in range(len(Open_Files)):
    Open_Files[line] = Open_Files[line].split(',')
Open_Files = np.array([np.array(i).astype(float) for i in Open_Files])

Emissions = []
for i in Files:
    Emissions.append(int(i[-10:-6].strip("_")))
print(Emissions)
```

```python
Iterator = int(len(Open_Files)/len(Files))
print(Iterator)
All_Emissions = np.zeros([len(Files),Iterator,2])
for i in range(len(Files)):
    All_Emissions[i] = Open_Files[Iterator*i:Iterator*(i+1)]
print(All_Emissions[0,:,1])
```

```python
for run in range(len(Emissions)):
    #print("_"*60)
    x_mindex = np.where(All_Emissions[run,:,1] == min(All_Emissions[run,:len(All_Emissions[run,:,1])//2,1]))[0][0]
    #print(x_mindex)
    random = 0
    inarow = 0
    for i in range(len(All_Emissions[run,:,1])):
        if abs(All_Emissions[run,i,1]) > 3e-3:
            x_maxdex = i-1
            break
    #print(x_maxdex)
    plt.plot(All_Emissions[run,x_maxdex:x_mindex,0], abs(All_Emissions[run,x_maxdex:x_mindex,1]/Emissions[run]*1000))
    #plt.gca().invert_yaxis()
plt.gca().invert_xaxis()
```
